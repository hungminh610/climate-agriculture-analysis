```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("climate_change_impact_on_agriculture_2024.csv")
print(df.head())
print(df.columns)
print(df.info())
print(df.isnull().sum())
df = df.dropna()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(10,8))
plt.show()
```

```python
top_countries = df.groupby('Country')['Economic_Impact_Million_USD'].sum().sort_values(ascending=False).head(10)

top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Economic Impact")
plt.xlabel("Country")
plt.ylabel("Million USD")
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight:
# Some countries experience significantly higher economic impact, suggesting unequal vulnerability to climate change.
```

```python
plt.figure()
plt.scatter(df['Average_Temperature_C'], df['Economic_Impact_Million_USD'])
plt.title("Temperature vs Economic Impact")
plt.xlabel("Temperature (°C)")
plt.ylabel("Economic Impact")
plt.show()
```

```python
# Insight:
# Higher temperatures tend to be associated with greater economic impact, indicating potential climate-related risks.
```

```python
# 1. Temperature Analysis
```

```python
import matplotlib.pyplot as plt

temp_trend = df.groupby('Year')['Average_Temperature_C'].mean()

plt.figure()
temp_trend.plot()
plt.title("Average Temperature Over Time")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Insight:
# The average temperature shows fluctuations over time, but there is a slight upward trend, especially in recent years.
# This may indicate the impact of climate change.
```

```python
# 2. Rainfall Analysis
```

```python
rain_trend = df.groupby('Year')['Total_Precipitation_mm'].mean()

plt.figure()
rain_trend.plot()
plt.title("Rainfall Over Time")
plt.xlabel("Year")
plt.ylabel("Precipitation (mm)")
plt.show()
```

```python
# Insight:
# Rainfall shows significant fluctuations over time with no clear trend.
# This indicates increasing variability in precipitation patterns, which may lead to both droughts and floods.
```

```python
# 3. Extreme Weather Events Analysis
```

```python
extreme = df.groupby('Year')['Extreme_Weather_Events'].sum()

plt.figure()
extreme.plot()
plt.title("Extreme Weather Events Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Events")
plt.show()
```

```python
# Insight:
# Extreme weather events show fluctuations over time, but there is a slight increase in frequency in recent years.
# This suggests growing climate instability and higher risk of disasters.
```

```python
# 4. Economic Impact Analysis
```

```python
impact = df.groupby('Year')['Economic_Impact_Million_USD'].sum()

plt.figure()
impact.plot()
plt.title("Economic Impact Over Time")
plt.xlabel("Year")
plt.ylabel("Million USD")
plt.show()
```

```python
# Insight:
# Economic impact has increased over time, indicating that climate-related events are causing greater financial damage.
```

```python
# 5. Conclusion
```

```python
# The analysis reveals that climate change is not only affecting environmental factors such as temperature and rainfall, but also increasing the frequency of extreme events.
# These changes are strongly associated with rising economic impact, highlighting the growing risks for agriculture and global economies.
```
